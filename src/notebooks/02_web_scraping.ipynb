```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = "https://en.wikipedia.org/wiki/Community_areas_of_Chicago"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url=url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
for table in soup.find_all("table"):
    print(table.get("class"))
```

```python
table = soup.find("table", class_="wikitable sortable plainrowheaders mw-datatable")
table
```

```python
for row in table.tbody.find_all("tr")[2:-1]:
    print(row)
    print("-" * 50)
```

```python
for row in table.tbody.find_all("tr")[2:-1]:
   celss = row.find_all("td")

   print(celss[0].get_text(strip=True))
```

```python
for row in table.tbody.find_all("tr")[2:-1]:
   header_cells = row.find_all("th")

   print(header_cells[0].get_text(strip=True))
```

```python
for row in table.tbody.find_all("tr")[2:-1]:
   celss = row.find_all("td")
   header_cells = row.find_all("th")

   print(celss[0].get_text(strip=True))
   print(header_cells[0].get_text(strip=True))
   print()
```

```python
data= []

for row in table.tbody.find_all("tr")[2:-1]:
   celss = row.find_all("td")
   header_cells = row.find_all("th")

   area_code = celss[0].get_text(strip=True)
   community_area_name =  header_cells[0].get_text(strip=True)
   
   data.append({"area_code": area_code, "community_are": community_area_name})
```

```python
data
```

```python
community_areas_df = pd.DataFrame(data)
community_areas_df["area_code"] = community_areas_df["area_code"].astype("int")
```
